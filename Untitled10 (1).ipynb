```python
import pandas as pd
import matplotlib.pyplot as plt
Data_Name= ["Yes_Fire","No_Fire"]
# Load the fire alarm data from the CSV file
data = pd.read_csv(r'D:\Projects\New folder\smoke_detection_iot.csv')
fire_alarm_data = data['Fire_Alarm']

# Calculate the frequency of different fire alarm events
alarm_counts = fire_alarm_data.value_counts()

# Create a pie chart to visualize the frequencies
plt.figure(figsize=(6, 6))
plt.pie(alarm_counts.values, labels=Data_Name, autopct='%1.1f%%', startangle=90)
plt.title('Fire Alarm Data')

plt.legend(title="Fire Alarm Events")

plt.show()
```
